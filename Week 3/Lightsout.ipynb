```python
n = 4  
cells = n * n
f = 0.9
eps = 1e-9
states = []
for num in range(2 ** cells):
    arr = [0] * cells
    x = num
    idx = cells - 1
    while x > 0:
        arr[idx] = x % 2
        x //= 2
        idx -= 1
    states.append(tuple(arr))
goal = states[0]
actions = list(range(cells))
mdp = {}
for state in states:
    mdp[state] = {}
    if state == goal:
        for a in actions:
            mdp[state][a] = [(1.0, state, 0, True)]
        continue
    for a in actions:
        flip = [a]
        r, c = divmod(a, n)

        if c + 1 < n:
            flip.append(a + 1)
        if c - 1 >= 0:
            flip.append(a - 1)
        if r + 1 < n:
            flip.append(a + n)
        if r - 1 >= 0:
            flip.append(a - n)

        next_state = list(state)
        for k in flip:
            next_state[k] = (next_state[k] + 1) % 2

        next_state = tuple(next_state)

        reward = 1 if next_state == goal else 0
        done = (next_state == goal)

        mdp[state][a] = [(1.0, next_state, reward, done)]
```

```python
V = {s: 0.0 for s in states}
V_new = {s: 0.0 for s in states}
policy = {s: 0 for s in states}
policy_old = {s: -1 for s in states}
V[goal] = 1.0
V_new[goal] = 1.0
while policy != policy_old:
    policy_old = policy.copy()
    while True:
        diff = 0.0
        for s in states:
            a = policy[s]
            val = 0.0
            for p, s2, r, _ in mdp[s][a]:
                val += p * (r + f * V[s2])
            diff = max(diff, abs(V[s] - val))
            V_new[s] = val

        V, V_new = V_new, V
        if diff < eps:
            break
    for s in states:
        best_val = -1e9
        best_act = 0
        for a in actions:
            q = 0.0
            for p, s2, r, _ in mdp[s][a]:
                q += p * (r + f * V[s2])
            if q > best_val:
                best_val = q
                best_act = a
        policy[s] = best_act
```

```python
limit = 100
m = 0
a1 = [tuple([1]*(n*n))]
for s in a1:
    if (s==goal):
        continue
    act = policy[s]
    tup = mdp[s][act][0]
    nexst = tup[1]
    movect = 1
    while(tup[3]==False):
        act = policy[tuple(nexst)]
        tup = mdp[tuple(nexst)][act][0]
        nexst = tup[1]
        movect+=1
        if (movect>limit):
            movect=0
            break
    m=max(movect, m)
print("Max steps : ",m)
```
